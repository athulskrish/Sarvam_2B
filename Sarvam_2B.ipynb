```python
from transformers import AutoModelForCausalLM, AutoTokenizer
import torch

# Load model and tokenizer
print("Loading Sarvam-1 model...")
model = AutoModelForCausalLM.from_pretrained("sarvamai/sarvam-1")
tokenizer = AutoTokenizer.from_pretrained("sarvamai/sarvam-1")

# Set device (CPU for your system with 8GB RAM)
device = torch.device("cpu")
model = model.to(device)

print("Model loaded successfully!")

# Example usage with Hindi text
text = "कर्नाटक की राजधानी है:"
print(f"Input text: {text}")

# Tokenize input
inputs = tokenizer(text, return_tensors="pt")

# Generate response
with torch.no_grad():  # Save memory
    outputs = model.generate(
        **inputs,
        max_new_tokens=50,
        temperature=0.7,
        do_sample=True,
        pad_token_id=tokenizer.eos_token_id
    )

# Decode and print result
result = tokenizer.decode(outputs[0], skip_special_tokens=True)
print(f"Generated text: {result}")

# Interactive loop for testing
print("\n--- Interactive Mode ---")
print("Enter text to complete (or 'quit' to exit):")

while True:
    user_input = input("\nEnter text: ")
    if user_input.lower() == 'quit':
        break

    inputs = tokenizer(user_input, return_tensors="pt")

    with torch.no_grad():
        outputs = model.generate(
            **inputs,
            max_new_tokens=30,
            temperature=0.7,
            do_sample=True,
            pad_token_id=tokenizer.eos_token_id
        )

    result = tokenizer.decode(outputs[0], skip_special_tokens=True)
    print(f"Completion: {result}")
```
